### Exercise4
Create a chart as shown below. Pay attention to the legends, axes and title, etc.

```python
# bar example
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.75       # the width of the bars: can also be len(x) sequence

#you code below
#hint: use plt.bar to create the chart
p1 = plt.bar(x=ind, height=menMeans, width=width, yerr=menStd)
p2 = plt.bar(x=ind, height=womenMeans, width=width, bottom='menMeans', yerr=womenStd)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'));
```

### Exercise 5
Given the precipitation for a city X, please create a chart that looks nice to you and can convey as much information as possible.

1) Please try at least three type of charts, including line, bar, scatter, etc.

2) You chart should have proper legends, axes names, ticks(use month name as X axis). 

```python
# create variables for each month of average precipitation for City X
jan = 0.70 * 25.4
feb = 0.75 * 25.4
mar = 1.85 * 25.4
apr = 2.93 * 25.4
may = 3.05 * 25.4
june = 2.02 * 25.4
july = 1.93 * 25.4
aug = 1.62 * 25.4
sept = 1.84 * 25.4
oct = 1.31 * 25.4
nov = 1.39 * 25.4
dec = 0.84 * 25.4

# you code here
# create a list of the converted monthly variables for the y-axis of your plot
precip = [jan, feb, mar, apr, may, june, july, aug, sept, oct, nov, dec] 

# create a list of the month names for the x-axis of your plot
months = ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# set plot size for the plot
plt.rcParams["figure.figsize"] = (8, 8)

# create the plot space upon which to plot the data
fig, ax = plt.subplots()

# add the x-axis and the y-axis to the plot
ax.plot(months, precip);

```

```python
# furthur improvement
# set plot size for all plots that follow
plt.rcParams["figure.figsize"] = (8, 8)

# set plot title size for all plots that follow
plt.rcParams['axes.titlesize'] = 20

# create the plot space upon which to plot the data
fig, ax = plt.subplots()

# add the x-axis and the y-axis to the plot
ax.plot(months, precip)

# set plot title
ax.set(title="Average Monthly Precipitation in City X")

# add labels to the axes
ax.set(xlabel="Month", ylabel="Precipitation (mm)");
```

```python
# use the bar chart

# set plot size for all plots that follow
plt.rcParams["figure.figsize"] = (8, 8)

# create the plot space upon which to plot the data
fig, ax = plt.subplots()

# add the x-axis and the y-axis to the plot
ax.bar(months, precip, color="green")

# set plot title
ax.set(title="Average Monthly Precipitation in City X")

# add labels to the axes
ax.set(xlabel="Month", ylabel="Precipitation (mm)");
```

```python
# scatter plot
plt.rcParams["figure.figsize"] = (8, 8)
fig, ax = plt.subplots()

# add the x-axis and the y-axis to the plot
ax.scatter(months, precip)
# set plot title
ax.set(title="Average Monthly Precipitation in City X")

# add labels to the axes
ax.set(xlabel="Month", ylabel="Precipitation (mm)");
```
